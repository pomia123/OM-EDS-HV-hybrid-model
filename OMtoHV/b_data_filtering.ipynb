```python
# =============================================================================
# 특정 column 기준 outlier filtering
# =============================================================================

import pandas as pd
import numpy as np
from scipy.stats import zscore

# --------------------------------------------------
# 제거 기준 column 직접 지정
# --------------------------------------------------

df = pd.read_csv(r'.\data\b_hv_with_features.csv')
df .fillna(0, inplace=True)

hv_cols = ['HV']
drop_cols = ['FILE_NAME'] + hv_cols
feat_cols = [c for c in df.columns if c not in drop_cols]

# --------------------------------------------------
# z-score 계산
# --------------------------------------------------

z_scores = np.abs(
    zscore(
        df[feat_cols],
        nan_policy='omit'
    )
)

# --------------------------------------------------
# threshold 설정
# --------------------------------------------------

threshold = 3

# --------------------------------------------------
# 모든 지정 column이 threshold 이하인 row만 유지
# --------------------------------------------------

mask = (z_scores < threshold).all(axis=1)

# --------------------------------------------------
# filtering 적용
# --------------------------------------------------

df_clean = (
    df[mask]
    .copy()
    .reset_index(drop=True)
)

# --------------------------------------------------
# 결과 출력
# --------------------------------------------------

df_clean.to_csv(r'.\data\c_cleaned_hv_with_features.csv', index=False)

print("=" * 60)
print(f"원본 데이터 수 : {len(df)}")
print(f"제거 후 데이터 수 : {len(df_clean)}")
print(f"제거된 데이터 수 : {len(df) - len(df_clean)}")
print("=" * 60)
```
